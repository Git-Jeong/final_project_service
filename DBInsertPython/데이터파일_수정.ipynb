```python
import pandas as pd
```

```python
data = pd.read_csv('./data/AirQualSample_0619_0625_s5.csv')
```

```python
data
```

```python
data.columns
```

```python
data = data.rename(columns={'TimeStamp': 'dtime'})
data = data[['dtime', 'Temp', 'Humidity', 'AtmosphericPress', 'PM1', 'PM2_5', 'PM10', 'CODen', 'CO2Den']]
```

```python
data
```

```python
data['PM1'] = data['PM1'].round().astype('Int64')
data['PM2_5'] = data['PM2_5'].round().astype('Int64')
data['PM10'] = data['PM10'].round().astype('Int64')
```

```python
data
```

```python
data['weekday'] = pd.to_datetime(data['dtime']).dt.strftime('%A')
```

```python
data
```

```python
data['weekday'].value_counts()
```

```python
data['time_hms'] = pd.to_datetime(data['dtime']).dt.time
```

```python
data['st_id'] = 48;
```

```python
data['CODen'].value_counts()
```

```python
try : 
    saveFileUrl = './data/v2_temp_csv.csv'
    data.to_csv(saveFileUrl, index=False)
    print(f'{saveFileUrl} 저장 완료. ')
except OSError as e :
    print(e)
```
